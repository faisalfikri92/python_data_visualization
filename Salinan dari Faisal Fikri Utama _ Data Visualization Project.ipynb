## Columns Renaming

```python
rename_mapper = {col:col.replace(" ", "_").lower() for col in df.columns}
df.rename(
    columns=rename_mapper,
    inplace=True
)

df.head()
```

## To Datetime datatype

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df[["order_date", "ship_date"]].head().info()
```

```python
df = df.sort_values(
    by=["order_id", "order_date", "ship_date"],
    ascending=[True, True, True],
    ignore_index=True
)

df.head(20)
```

## Remove "$" in ["sales", "profit", "shipping_cost" ]

```python
dollar_cols = ["sales", "profit", "shipping_cost"]
for col in dollar_cols:
  df[col] = (
      df[col]
      .str.replace("[^\d.]","", regex =True)
  )
  df.rename(columns={col: col+"($)"}, inplace=True)
print("Formatting & Renaming, Done!")
```

## Monthly Sales Trends

```python
monthly_sum_sales = df.groupby("months").agg({"sales($)": "sum"})
monthly_sum_sales
```

```python
ordered_months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

Month_CI = pd.CategoricalIndex(
    monthly_sum_sales.index,
    ordered=True,
    categories=ordered_months
)

Month_CI
```

```python
monthly_sum_sales.index = Month_CI
monthly_sum_sales = monthly_sum_sales.sort_index()
monthly_sum_sales
```

```python
import matplotlib.ticker as ticker
```

```python
# Create line chart for monthly sales
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(
    monthly_sum_sales.index,
    monthly_sum_sales["sales($)"],
    marker='*',
    mfc='red', # marker face color
    mec='red', # marker edge color
    linestyle='-'
)

ax.set_title('Monthly Sales Trends')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.grid(True)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))

# Show the line chart
fig.show()
```

## 2.&nbsp; **Product Sales and Profitability**

```python
product_profit = df.groupby('product').agg({'profit($)':'sum'}).sort_values(by='profit($)', ascending=False)
```

```python
product_sales = df.groupby('product').agg({'sales($)':'sum'}).sort_values(by='sales($)', ascending=False)
```

```python
top_10_products = product_sales.nlargest(10, 'sales($)')
top_10_profit = product_profit.nlargest(10, 'profit($)')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(top_10_products.index,
      top_10_products['sales($)'],
      color='orange',
      label='Sales')
plt.xticks(rotation=45)

ax.scatter(top_10_products.index,
           top_10_profit['profit($)'],
           color='blue',
           marker='o',
           s=250,
           label='Profit')

ax.set_title("Products Sales and Profit")
ax.set_xlabel("Product")
ax.set_ylabel("Sales ($)")
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x / 1000:.0f}K'))
ax.legend()
fig.tight_layout()
plt.show()
```
